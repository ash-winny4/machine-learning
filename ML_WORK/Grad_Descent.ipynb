```python
import numpy as np
```

```python
def gradient_descent(x,y):
    mcurr = bcurr = 0
    lrate = 0.07
    iterations = 1000
    n = len(x)

    for i in range(iterations):
        ypred = mcurr*x + bcurr
        cost = (1/n)*sum([val**2 for val in (y-ypred)])
        m = -(2/n)*sum(x*(y-ypred))
        b = -(2/n)*sum(y-ypred)
        mcurr = mcurr - lrate*m
        bcurr = bcurr - lrate*b
        print("weight: {}, bias: {}, cost: {}, iteration: {}".format(mcurr,bcurr,cost,i))


x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
gradient_descent(x,y)
```
