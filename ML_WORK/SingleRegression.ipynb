```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("D:\Datasets\home_prices.csv")
df
```

```python
plt.scatter(df.area_sqr_ft,df.price_lakhs)
plt.xlabel('area')
plt.ylabel('price in INR')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area_sqr_ft']],df.price_lakhs)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
